```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Sales']=pd.to_numeric(df['Sales'])
df['City'] = df['City'].astype(str)
```

```python
total_sales=df.groupby(by='City')['Sales'].sum().reset_index()
total_sales
```

```python
total_sales = total_sales[total_sales['City'] != 'nan']
```

```python
plt.figure(figsize=(10, 6))  # Optional: Adjust figure size
plt.bar(total_sales['City'], total_sales['Sales'])
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.title('Total Sales by City')
plt.xticks(rotation=45)  # Optional: Rotate x-axis labels for better readability
plt.show()
```

```python
Codes ={ 'Atlanta': 'ATL', 'Austin': 'AU','Boston': 'BO', 'Dallas': 'DA', 'Los Angeles': 'LA', 'New York City': 'NY', 'Portland': 'PORT', 'San Francisco': 'SF',
'Seattle': 'SA' }
```

```python
# Ensure 'City' column has no extra spaces
df['City'] = df['City'].str.strip()

# Map 'City' to 'City Code' using the dictionary
df['City_Codes'] = df['City'].map(Codes)
```

```python
df.head(1)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.set_index('Order Date',inplace=True)
```

```python
df.dropna(subset=['Sales'], inplace=True)
```

```python
grouped = df.groupby('City')
```

```python
result = {}
for city, group in grouped:
    # Resample the data into six-month intervals and sum up 'Sales'
    resampled = group.resample('6M').sum()['Sales']
    result[city] = resampled

# Convert the result to a DataFrame for better readability
resampled_df = pd.concat(result, axis=1)

print(resampled_df)
```
